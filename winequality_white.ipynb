# Data Visualization

# To check missing values

```python
sns.heatmap(df.isnull(),cbar=False,yticklabels=False,cmap = 'viridis')
```

- Dataset has no missing values.


# To check correlation

```python
plt.figure(figsize=(9,6))
sns.heatmap(df.corr(),cmap='Blues', annot=True) 
```

- Dark shades represents positive correlation while lighter shades represents negative correlation.


```python
#Quality correlation matrix
k = 12 #number of variables for heatmap
cols = df.corr().nlargest(k,'quality')['quality'].index
cm = df[cols].corr()
plt.figure(figsize=(10,6))
sns.heatmap(cm, annot=True, cmap = 'RdYlGn')
```

- Here we can infer that "density" has strong positive correlation with "residual sugar" whereas it has strong negative correlation with "alcohol".
- "free sulphur dioxide" and "citric acid" has almost no correlation with "quality"
- Since correlation is zero we can infer there is no linear relationship between these two predictors.However it is safe to drop these features in case future applying Linear Regression model to the dataset. 

# Checking Outliers

```python
l = df.columns.values
number_of_columns=12
number_of_rows = 12

plt.figure(figsize=(number_of_columns,5* number_of_rows))
for i in range(0,len(l)):
    plt.subplot(number_of_rows  ,number_of_columns,i+1)
    sns.set_style('whitegrid')
    sns.boxplot(y =df[l[i]],color='green', orient ='v')
    plt.tight_layout()
```

# Checking distribution-Skewness

```python
plt.figure(figsize=(2* number_of_rows,number_of_columns))
for i in range(0,len(l)):
    plt.subplot(1,number_of_columns,i+1)
    sns.distplot(df[l[i]],hist=True ,kde=False) 
```

- "pH" column appears to be normally distributed
- remaining all independent variables are right skewed/positively skewed.   

```python
x= df["residual sugar"]
y= df["density"]
plt.plot(x,y,'o')
m, b = np.polyfit(x, y, 1)
plt.plot(x, m*x + b)
```

- Linear regression model for correlation between residual sugar and density columns
